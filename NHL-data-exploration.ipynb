```python
# it may be easier to create a copy of our df, and then organize our variables there. this allows to keep the naming structure of 'stat.XXXXX'
gamelog_df = pd.json_normalize(my_guy.game_log)
#numarically encoded dataframe
encoded_df = gamelog_df
encoded_df['stat.didScore'] = encoded_df['stat.goals'].apply(lambda x: 1 if x != 0 else 0)
encoded_df['stat.timeOnIce'] = encoded_df['stat.timeOnIce'].apply(lambda x: int(x.split(':')[0]) * 60 + int(x.split(':')[1]))
encoded_df['stat.evenTimeOnIce'] = encoded_df['stat.evenTimeOnIce'].apply(lambda x: int(x.split(':')[0]) * 60 + int(x.split(':')[1]))
encoded_df['stat.powerPlayTimeOnIce'] = encoded_df['stat.powerPlayTimeOnIce'].apply(lambda x: int(x.split(':')[0]) * 60 + int(x.split(':')[1]))
encoded_df['stat.shortHandedTimeOnIce'] = encoded_df['stat.shortHandedTimeOnIce'].apply(lambda x: int(x.split(':')[0]) * 60 + int(x.split(':')[1]))
encoded_df.head()
```

```python
#selecting the variables of value for a goal prediction: the "is" columns, the "stat" columns, 
# and "opponent.name" which we will oneHotencode into binary representations
# 
selected_columns = encoded_df.filter(regex='^(is|stat|opponent.name)')
selected_columns.head()
```

```python
#one hot encode for the opponent.name variable
selected_one_hot = pd.get_dummies(selected_columns, columns=['opponent.name'])
#cast all columns to float64 type
selected_one_hot = selected_one_hot.astype('float64')
selected_one_hot.head()
```

```python
# we have to drop data for number of goals and points, these are redundant-related data that will unfairly influence our prediction.
selected_one_hot = selected_one_hot.drop(['stat.goals','stat.points'], axis=1)
```

```python
selected_one_hot.shape
```

```python
#lets do some ML, we need some tools. 

# Preprocessing allows us to standarsize our data
from sklearn import preprocessing
# Allows us to split our data into training and testing data
from sklearn.model_selection import train_test_split
# Allows us to test parameters of classification algorithms and find the best one
from sklearn.model_selection import GridSearchCV
# Logistic Regression classification algorithm
from sklearn.linear_model import LogisticRegression
# Support Vector Machine classification algorithm
from sklearn.svm import SVC
# Decision Tree classification algorithm
from sklearn.tree import DecisionTreeClassifier
# K Nearest Neighbors classification algorithm
from sklearn.neighbors import KNeighborsClassifier
```

```python
#lets impute some values for NaN
#and we need to see which variables contain NaN values
print(selected_one_hot.isna().any())
```

```python
y = selected_one_hot['stat.didScore']
x = selected_one_hot.drop(['stat.didScore'],axis=1)
```

```python
#for shooting percentage
x['stat.shotPct'] = x['stat.shotPct'].fillna(x['stat.shotPct'].mean())
```

```python
#for faceoff percentage, a little cleaner code

x['stat.faceOffPct'].fillna(x['stat.faceOffPct'].mean(), inplace=True)
x.head()
```

```python

# pca = PCA(n_components=2)

# # Fit the PCA object to the data and transform the data
# x_pca = pca.fit_transform(x)

# # Back into a single dataframe
# data_pca = pd.DataFrame(data=x_pca, columns=['PC1', 'PC2'])


# # Print the explained variance ratio for each principal component
# print(pca.explained_variance_ratio_)
```

```python
#starting by selecting what we want to predict for
Y = y.to_numpy()
Y.shape
```

```python
transform = preprocessing.StandardScaler()
X = transform.fit_transform(x)
X.shape
```

```python
#split the data for training and testing
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
#lets see how many test samples we have
Y_test.shape, X_test.shape
```

```python
LR_model = LogisticRegression(random_state=42)
```

```python
param_grid = {
    'solver': ['newton-cg', 'lbfgs', 'liblinear'],
    'penalty': ['l1', 'l2', 'elasticnet'],
    'C': [0.01, 0.1, 1, 10, 100],
}


LR_gridsearch = GridSearchCV(LR_model, param_grid, cv=10)
LR_gridsearch.fit(X_train, Y_train)
print(LR_gridsearch.best_params_)
print("accuracy: ", LR_gridsearch.best_score_)
```

```python
def plot_confusion_matrix(y,y_predict):
    "this function plots the confusion matrix"
    from sklearn.metrics import confusion_matrix

    cm = confusion_matrix(y, y_predict)
    ax= plt.subplot()
    sns.heatmap(cm, annot=True, ax = ax); #annot=True to annotate cells
    ax.set_xlabel('Predicted labels')
    ax.set_ylabel('True labels')
    ax.set_title('Confusion Matrix'); 
    ax.xaxis.set_ticklabels(['did not score', 'score']); ax.yaxis.set_ticklabels(['did not score', 'scored'])
```

```python
print("accuracy of the model on test reserved data: ", LR_gridsearch.score(X_test,Y_test))
yhat=LR_gridsearch.predict(X_test)
plot_confusion_matrix(Y_test,yhat)
```
