```python
#lets create a categorial variable for if the player scored in each game
#we then append values of that variable to the dataframe

features_df = gamelog_df
features_df['didScore'] = features_df['stat.goals'].apply(lambda x: 1 if x != 0 else 0)
features_df['didScore'].value_counts()
```

```python
features_df.plot(y='didScore', title = 'didScore, for each game', legend=True, xlabel='Game #', ylabel='didScore')
```

```python
#feature engineer a gameCount variable to use in a scatter plot

features_df['gameCount'] = features_df.reset_index().index + 1
features_df.head()
```

```python
import seaborn as sns
sns.catplot(data=features_df, x='gameCount',y='didScore',aspect=5)
```

```python
# features_df['iceTime'] = pd.to_timedelta(features_df['stat.timeOnIce']).dt.total_seconds()
# convert time measurement to seconds for plotting, appending as an engineered data point to our dataframe
features_df['iceTime'] = features_df['stat.timeOnIce'].apply(lambda x: int(x.split(':')[0]) * 60 + int(x.split(':')[1]))
```

```python
features_df[['stat.timeOnIce','iceTime']].head()
```

```python
sns.scatterplot(data=features_df[features_df['iceTime']>=500], x='gameCount', y='iceTime',hue='didScore')
```

```python
#games with more than 500 seconds of iceTime
time_filter_df = features_df[features_df['iceTime']>=500]
```

```python
# create two scatter plots with different colors for didScore=True and didScore=False
sns.scatterplot(data=time_filter_df[time_filter_df['didScore']==1], x='gameCount', y='iceTime', color='orange', label='didScore=True')
sns.scatterplot(data=time_filter_df[time_filter_df['didScore']==0], x='gameCount', y='iceTime', color='blue', label='didScore=False')

# add best fit line for each scatter plot
sns.regplot(data=time_filter_df[time_filter_df['didScore']==1], x='gameCount', y='iceTime', scatter=False, color='black')
```

```python
#check to see what percentage of values are null
null_pct = features_df.isnull().sum() / features_df.count() * 100
null_pct[features_df.count() == 0] = 0
null_pct
```
